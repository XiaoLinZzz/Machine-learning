```python
import pandas as pd
```

```python
df = pd.read_csv(r"0\0_left.csv")
```

```python
df.head()
```

```python

attributes = pd.Series(df.columns.tolist()[2:]).apply(lambda x: x.strip()[:-2]).drop_duplicates().reset_index(drop ='True')
attributes
```

```python
from itertools import permutations 
  
perm = permutations(attributes, 2) 
perm = list(perm)
```

```python
for i in range(len(perm)):
    x1 = df[" "+perm[i][0]+"_X"]
    y1 = df[" "+perm[i][0]+"_Y"]
    z1 = df[" "+perm[i][0]+"_Z"]
    x2 = df[" "+perm[i][1]+"_X"]
    y2 = df[" "+perm[i][1]+"_Y"]
    z2 = df[" "+perm[i][1]+"_Z"]
    df[perm[i]]=((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)**1/2
    
```
